```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
plt.rcParams["figure.figsize"] = (12,4)
sns.set_style("whitegrid")

# importing libraries and setting up matplotlib and seaborn
```

```python
path = "C:\\Users\\UserPC\\Desktop\\my_folder\\Learning alongside DProf\\kaggle and LinkedIn\\datasets\\Superstore-Sales.csv"
data = pd.read_csv(path, index_col = "Order Date", encoding = "cp1252", parse_dates = True)
data.head()

# loading the data, setting the index and parsing dates, checking the first few rows
```

```python
data["Order Quantity"].plot() # Select and plot the "Order Quantity" column
data = data.sort_index()  # Sort by Order Date
plt.xlabel("Order Date") # Label the x-axis
plt.ylabel("Order Quantity") # Label the y-axis
plt.title("Order Quantity Over Time (Full Data)") # Add a title
plt.show() # Show the plot

```

```python
time_data = data.sample(n=100, random_state=25, axis=0) # Randomly sample 100 rows
time_data = time_data.sort_index()  # Sort by Order Date
time_data["Order Quantity"].plot() # Plot the sampled data
plt.xlabel("Order Date") # Label the x-axis
plt.ylabel("Order Quantity") # Label the y-axis
plt.title("Order Quantity Over Time (Sampled Data)") # Add a title
plt.show() # Show the plot
```
