```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {"names" : ["Steve", "John", "Richard", "Sarah", "Randy", "Michael", "Julie"],
        "ages" : [20, 22, 20, 21, 24, 23, 22],
        "gender" : ["male", "male", "male", "female", "male", "male", "female"],
        "rank" : [2, 1, 4, 5, 3, 7, 6]}

df = pd.DataFrame(data)
df

# creating a basic data set
```

<h4>Matplotlib - Barchart</h4>

```python
plt.bar(df["names"], df["ages"])
plt.xlabel("Names")
plt.ylabel("Ages")
plt.title("Age of Staff Members")
plt.show()

# basic bar chart using matplotlib
```

<h4>Seaborn - Barchart</h4>

```python
plot = sns.barplot(data = df, x = "names", y = "ages")
plot.set_title("Age of Staff Members")
plot

# basic bar chart using seaborn
```

<h4>Matplotlib - Linechart</h4>

```python
plt.plot(df["names"], df["ages"])
plt.xlabel("Names")
plt.ylabel("Ages")
plt.title("Age of Staff Members")
plt.show()

# basic line chart using matplotlib
```

<h4>Seaborn - Linechart</h4>

```python
plot = sns.lineplot(data = df, x = "names", y = "ages")
plot.set_title("Age of Staff Members")
plot
```

<h4>Matplotlib - Piechart</h4>

```python
plt.pie(df["ages"], labels = df["names"])
plt.title("Age of Staff Members")
plt.show()

#basic pie chart using matplotlib
```

<h4>Seaborn - Piechart</h4>

```python
colours = sns.color_palette("pastel")[0:5]
plt.pie(df["ages"], labels = df["names"], colors = colours)
plt.title("Age of Staff Members")
plt.show()

# basic pie chart using matplotlib with seaborn colors
```
