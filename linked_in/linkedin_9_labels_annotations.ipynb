```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
plt.rcParams["figure.figsize"] = (12,4)

# importing libraries and setting up matplotlib
```

```python
x = range(1,10)
y = [1, 2, 3, 4, 0.5, 4, 3, 2, 1]
plt.bar(x, y)
plt.xlabel("X axis label")
plt.ylabel("Y axis label")
plt.title("Bar Chart Example")
plt.show()

# labeling the axes and the chart
```

```python
z = [1, 2, 3, 4, 0.5]
veh = ["bicycle", "motorbike", "car", "van", "stroller"]
plt.pie(z)
plt.legend(veh, loc = "lower right")
plt.title("Pie Chart Example")
plt.show()

# pie chart with labels
```

```python
file_path = "C:\\Users\\UserPC\\Desktop\\my_folder\\Learning alongside DProf\\kaggle and LinkedIn\\datasets\\mtcars.csv"
data = pd.read_csv(file_path)
data.rename(columns = {data.columns[0]: "car_name"}, inplace = True) 

# reading in the data and renaming the first column

mpg = data["mpg"]
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 1, 1])

# creating the figure and axes objects

mpg.plot()

# plotting the data

ax.set_xticks(range(len(data)))
ax.set_xticklabels(data["car_name"], rotation = 75)
ax.set_ylabel("Miles per Gallon")
ax.set_title("Miles per Gallon for Various Car Models")

# labeling the axes and the chart

ax.set_ylim(0, mpg.max() + 5)

# setting the y-axis limits

plt.show()

# displaying the plot
```
