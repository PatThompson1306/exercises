```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
plt.rcParams["figure.figsize"] = (5,3)

# importing libraries and setting up matplotlib
```

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

# defining x and y data for plotting (can be taken from a dataframe column)

fig = plt.figure()  # create a figure object
ax = fig.add_axes([.1,.1,1,1])  # add axes to the figure
ax.plot(x,y)  # plot on those axes
plt.show()  # display the plot

# creating a figure and axes, plotting on the axes and displaying the plot
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)  # add 2 subplots to the figure (one row, two columns)
ax1.plot(x)  # plot on the first axes
ax2.plot(x, y)  # plot on the second axes
plt.show()  # display the plot

# creating subplots and plotting on them
```

<h4>Plot styling</h4>

```python
x = range(1, 10)
y = [1, 2, 3, 4, 0.5, 4, 3, 2, 1]
plt.bar(x, y)
plt.show()
```

```python
width = [0.5, 0.5, 0.5, 0.9, 0.9, 0.9, 0.5, 0.5, 0.5]
color = "salmon"
plt.bar(x, y, width = width, color = color, align = "center")
plt.show()

# creating a bar chart with custom width, color and alignment
```
