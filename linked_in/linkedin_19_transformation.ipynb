<h4>Normalisaion & Standardisation</h4>



Normalisation and standardisation are data preprocessing techniques that play a crucial role in data analysis and machine learning.

Normalisation is the process whereby data values are scaled to fall within a range of 0 - 1.  It maintains the original distribution of a data set without altering their ranges.

Standardisation is th eprocess which scales data so it has a mean value of 0 and a standard deviation of 1.

There is no strict rule in applying raw, normalised or standardised data - the data mining / exploration process encourages use of all to define the best outputs.

They matter because - 

<h5>Different feature scales distort models</h5>

If one feature is in the thousands (e.g., income) and another is between 010 (e.g., ratings), algorithms that rely on distance (like k-NN, clustering, SVM) or gradient descent (like linear/logistic regression, neural networks) will be biased toward the larger-scaled feature.

<h5>Faster and more stable training</h5>

Gradient descent converges faster when features are scaled similarly. Without scaling, optimisation can "zig-zag" and take much longer.

<h5>Improved interpretability of coefficients</h5>

Standardised features let you compare feature importance more fairly, since theyre on the same scale.

<h5>Better performance in distance-based models</h5>

Models like k-NN, k-means, and PCA are very sensitive to magnitude differences. Scaling ensures all variables contribute proportionally.

<h5>Helps regularisation techniques</h5>

Lasso and Ridge regression penalise coefficients; if features arent scaled, some coefficients are penalised unfairly.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, scale 
```

```python
file_path = "C:\\Users\\UserPC\\Desktop\\my_folder\\Learning alongside DProf\\kaggle and LinkedIn\\datasets\\mtcars.csv"
data = pd.read_csv(file_path)
```

<h4>Normalisation</h4>

```python
plt.plot(data['mpg']) #  here we can see that the data is not normalised
plt.title('Miles per Gallon before Normalisation')
```

```python
minmax_scaler = MinMaxScaler()  # create the normalisation object
minmax_scaler.fit(data[['mpg']]) # fit the data to the normalisation object
scaled_data = minmax_scaler.transform(data[['mpg']]) # transform the data using the normalisation object
plt.plot(scaled_data) #  plot the normalised data - the Y axis is now between 0 and 1 as opposed to 0 and 35 pre normalisation
plt.title('Miles per Gallon after Normalisation')
```

<h4>Standardisation</h4>

```python
standardised_data = scale(data[['mpg']]) # standardise the data
plt.plot(standardised_data) #  plot the standardised data - the Y axis is now between -3 and +3 as opposed to 0 and 35 pre normalisation
plt.title('Miles per Gallon after Standardisation')
```
