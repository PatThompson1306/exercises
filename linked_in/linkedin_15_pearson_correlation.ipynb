```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from pylab import rcParams
%matplotlib inline
rcParams['figure.figsize'] = 8, 4
sns.set_style('whitegrid')

path = r"C:\\Users\\UserPC\Desktop\\my_folder\\Learning alongside DProf\\kaggle and LinkedIn\datasets\\mtcars.csv"
data = pd.read_csv(path)
data.rename(columns = {data.columns[0]: "car_name"}, inplace = True) 

# importing libraries, defining the style presets and loading the data
```

```python
sns.pairplot(data)
plt.show()

# creating a pairplot to visualize relationships between variables
```

```python
x = data[["mpg", "hp", "qsec", "wt"]]
sns.pairplot(x)
plt.show()

# creating a pairplot to visualize relationships between selected variables
```

```python
mpg = data["mpg"]
hp = data["hp"]
qsec = data["qsec"]
wt = data["wt"]

# defining variables for correlation analysis

pearsonr_coefficient, p_value = stats.pearsonr(mpg, hp)
print(f"Pearson's correlation coefficient: {pearsonr_coefficient.round(3)}")
print(f"P-value: {p_value.round(3)}")

# calculating Pearson's correlation coefficient and p-value between mpg and hp
```

```python
pearsonr_coefficient, p_value = stats.pearsonr(mpg, qsec)
print(f"Pearson's correlation coefficient: {pearsonr_coefficient.round(3)}")
print(f"P-value: {p_value.round(3)}")

# calculating Pearson's correlation coefficient and p-value between mpg and qsec
```

```python
pearsonr_coefficient, p_value = stats.pearsonr(mpg, wt)
print(f"Pearson's correlation coefficient: {pearsonr_coefficient.round(3)}")
print(f"P-value: {p_value.round(3)}")

# calculating Pearson's correlation coefficient and p-value between mpg and wt
```

```python
corr = x.corr()
print(corr)

# a quick way to use pandas to calculate the correlation matrix for selected variables
```

```python
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels = corr.columns.values, annot = True, fmt = ".2f")
plt.show()

# creating a heatmap to visualize the correlation matrix using seaborn
```
