<h4>ML Model Pipeline</h4>

At a very high level, an ML pipeline looks like -





```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
path = "C:\\Users\\UserPC\\Desktop\\my_folder\\Learning alongside DProf\\kaggle and LinkedIn\\datasets\\iris.csv"
df = pd.read_csv(path)
```

<h4>Separating Features & Labels</h4>

```python
df.Species.unique() # check the unique values in the target column
X = df.iloc[:,1:5] # remove the Species column (this will be our label / y variable)
print(X) # X represents our features / independent variables
```

```python
y = df.iloc[:,5] # this is our label / y variable
print(y) # y represents our target / dependent variable
```

<h4>Train Test Split</h4>

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0) # creating a 70% training and 30% test data split
```

<h4>Training Decision Tree Classifier</h4>

```python
clf = DecisionTreeClassifier() # creating a Decision Tree Classifier
clf.fit(X_train,y_train) # training the model
```

```python
y_pred = clf.predict(X_test) # making predictions on the test set
print(y_pred)
```

<h4>Evaluation Metrics</h4>

```python
accuracy = metrics.accuracy_score(y_test, y_pred) # calculating the accuracy of the model by comparing the actual values (y_test) with the predicted values (y_pred)
print("Accuracy:", accuracy)
```
