<h4>Concatenating / Transforming</h4>

```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame(np.arange(36).reshape(6, 6))
data_two = pd.DataFrame(np.arange(15).reshape(5, 3))
print(data)
print("---")
print(data_two)

# creating a couple of new sample dataframes
```

<h4>Concatenating Data</h4>

```python
pd.concat([data, data_two], axis = 1)

# concatenating along the horizontal axis (adding more columns)
```

```python
pd.concat([data, data_two])

# concatenating along the vertical axis (adding more rows)
```

<h4>Dropping Data</h4>

```python
data
```

```python
drop_rows = data.drop([0, 2])
drop_rows
# dropping rows 0 and 2
```

```python
drop_cols = drop_rows.drop([0, 2], axis = 1)
drop_cols
# dropping columns 0 and 1 (difference being changing the axis parameter)
```

<h4>Adding Data</h4>

```python
more_data = pd.Series(np.arange(6))
more_data.name = "moar_data!"
more_data

# creatng some more data to add to original data
```

```python
new_data = data.join(more_data, how="outer")
new_data

# joining the new data to the original data frame
```

<h4>Sorting Data</h4>

```python
df_sorted = data.sort_values(by = 5, ascending = False)
df_sorted

# sorting the data frame by column 5 in descending order
```
