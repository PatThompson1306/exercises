```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
plt.rcParams["figure.figsize"] = (10,5)
sns.set_style("whitegrid")


file_path = "C:\\Users\\UserPC\\Desktop\\my_folder\\Learning alongside DProf\\kaggle and LinkedIn\\datasets\\mtcars.csv"
data = pd.read_csv(file_path)
data.rename(columns = {data.columns[0]: "car_name"}, inplace = True) 
data.index = data["car_name"]
mpg = data["mpg"]

# importing libraries and reading data 
```

```python
# Matplotlib histogram

mpg.plot(kind = 'hist')  # Create a histogram of the "mpg" column
plt.show() # Show the plot
```

```python
# Seaborn distribution plot

sns.displot(mpg) # Create a distribution plot of the "mpg" column
plt.show() # Show the plot
```

```python
# Matplotlib scatter plot

data.plot(kind = "scatter", x = "hp", y = "mpg", c = "darkgray", s = 150) # Scatter plot of "hp" vs "mpg"
plt.show() # Show the plot
```

```python
# Seaborn scatter plot

sns.regplot(x = "hp", y = "mpg", data = data, scatter = True) # Scatter plot with regression line
plt.show() # Show the plot
```

```python
# Seaborn scatter plot matrix

sns.pairplot(data[["mpg", "disp", "hp", "wt"]]) # Pairplot of selected variables colored by "cyl"
plt.show() # Show the plot
```

```python
# Box plot using Matplotlib

data.boxplot(column = "mpg", by = "am")
data.boxplot(column = "wt", by = "am")
plt.show()
```

```python
# Box plot using Seaborn

sns.boxplot(x = "am", y = "mpg", data = data, hue = "am", palette = "hls")
plt.show()
```
