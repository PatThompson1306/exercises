```python
flines = "Hello.  How are you?  I am fine."
glines = "How are you Pazza?  I am fine.  Banana!"
d = difflib.Differ() # Create a Differ object
diff = d.compare(flines.split(), glines.split()) # Compare the two lines word by word

for line in diff: # Print the differences
    print(line)
```

This code uses Python's difflib library to compare two strings word by word and show the differences between them. Here's what's happening:

<h5>Input strings:</h5>

flines: "Hello. How are you? I am fine."
glines: "How are you Pazza? I am fine. Banana!"
The difflib.Differ() object:

Creates a tool to compare sequences and generate human-readable differences
The compare() method splits both strings into words and compares them.

<h5>Output symbols meaning:</h5>

- (- minus): Words that exist in the first string (flines) but NOT in the second string (glines)
- (+ plus): Words that exist in the second string (glines) but NOT in the first string (flines)
- No symbol: Words that are identical in both strings
- (? question mark): Indicates a line with detailed character-by-character differences

<h5>Breaking down the output:</h5>

- (- Hello). - "Hello." appears only in the first string
- How are - These words are common to both strings
- (- you?) - "you?" appears only in the first string
+ (+ you + Pazza?) - "you" and "Pazza?" appear only in the second string
- I am fine. - These words are common to both strings
+ (+ Banana!) - "Banana!" appears only in the second string
