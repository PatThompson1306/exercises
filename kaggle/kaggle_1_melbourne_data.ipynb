```python
import pandas as pd 
pd.options.display.float_format = '{:,.2f}'.format # rounding float values to 2 decimal places
```

```python
file_path = "C:\\Users\\UserPC\\Desktop\\my_folder\\Learning alongside DProf\\kaggle and LinkedIn\\datasets\\melb_data.csv"
melb_data = pd.read_csv(file_path)
melb_data.describe()

# descriptive stats immediately identify 13,580 rows of data for most columns aside from three
```

```python
melb_data.columns

# viewing the column headers
```

```python
melb_data = melb_data.dropna(axis=0)
melb_data.describe()

# this drops data rows where there is a missing data point in the row.  This is crude but serves our purpose here!
# it results in 6,196 rows of complete data
```

```python
y = melb_data.Price

# this is the "target variable" we want to predict.
# next we need to identify the features whic we will use to make the prediction
```

```python
features = ["Rooms", "Bathroom", "Landsize", "BuildingArea",
            "YearBuilt", "Lattitude", "Longtitude"]
X = melb_data[features]

# here we're selecting our features (independent variables) for the model
# we're then placing the features into a variable called X
```

```python
from sklearn.tree import DecisionTreeRegressor # type: ignore
model = DecisionTreeRegressor(random_state = 1)
model.fit(X, y)

# here we have imported a decision tree regressor model from scikit-learn and instantiated it
# we have also fitted the model to our training data
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are:")
print(model.predict(X.head()))

# here we have made predictions for the first 5 houses in our dataset
```

```python
from sklearn.metrics import mean_absolute_error # type: ignore

predicted_home_values = model.predict(X)
mae = mean_absolute_error(y, predicted_home_values)
print("The mean absolute error of our predictions is:")
print(f"${round(mae, 2)}")

# here we have calculated the mean absolute error (MAE) of our predictions
# the MAE is a measure of how far off our predictions are from the actual values
# it is calculated as the average of the absolute differences between predicted and actual values
# in this instance however the MAE is established from the training data - so there is no real measure of performance on unseen data
```

```python
from sklearn.model_selection import train_test_split # type: ignore

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
model = DecisionTreeRegressor()
model.fit(train_X, train_y)

# here we have split our data into training and validation sets
# we then fit our model to the training data

validated_predictions = model.predict(val_X)

print("Making predictions for the first five 5 houses in the validation data set:")
print("The predictions are:")
print(model.predict(val_X.head()))
print("The actual values are:")
print(val_y.head().tolist()) # converting the head() display to a list
print("The mean absolute error of our validated predictions is:")
print(f"${round(mean_absolute_error(val_y, validated_predictions), 2)}")

# once we have trained the model we can use it to make predictions on the validation set
# we can see that the MAE is huge and clearly indicates that the model is not performing well
```

Thus far we have taken data, removed nul values, identified the target variable, identified the features,split the data into training and validation sets, defined an initial training model, applied this to the data annd then used Mean Absolute Error to establish the model's accuracy.

We are now going to tweak this model using variations of leaf node to establish if we can improve the MEA score and thereby make the model more accurate.

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return mae

# defines a function to find the best number of leaf nodes 
# parameters: max_leaf_nodes, train_X, val_X, train_y, val_y are applied
# the model is defined and trained and validated with MAE scores returned based on the varied number of leaf nodes
```

```python
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print(f"Max leaf nodes: {max_leaf_nodes}\t\tMean Absolute Error: {round(my_mae, 2)}")

# here we have taken the function and looped over four alternative leaf nodes resulting in varying MAE scores.
```
